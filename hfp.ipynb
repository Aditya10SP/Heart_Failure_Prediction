```python
# importing basic libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

**TASK 1**

```python
df = pd.read_csv("/content/drive/MyDrive/Heart Failure Prediction/heart.csv")
df.head()
```

**ONE HOT ENCODING**

```python
##One-hot encoding

df2 = pd.read_csv("/content/drive/MyDrive/Heart Failure Prediction/heart.csv", usecols=['Sex', 'ChestPainType', 'RestingECG', 'ExerciseAngina', 'ST_Slope'])
df2.head()
```

```python
for col in df2.columns:
  print(col,"-", len(df2[col].unique()),"unique categories in this feature")
```

```python
df3=pd.get_dummies(df2)
df3.shape
```

```python
df=pd.concat([df, df3], axis='columns')
df.head()
```

**FINAL DATA FRAME AFTER ONE-HOT ENCODING**

```python
df=df.drop(df2, axis='columns')
df.head()
```

**NORMALISING THE DATAFRAME**

```python
for col in df.columns:
  if df[col].all().max() > 5:
    df[col] = df[col]/df[col].mean()

df.head()
```

**APPLYING THE LOGISTIC REGRESSION MODEL**

```python
X_columns = []
for col in df.columns:
  if col != 'HeartDisease':
    X_columns.append(col)
```

```python
X = df[X_columns]
y = df['HeartDisease']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
clf = LogisticRegression()
clf.fit(X_train, y_train)
```

```python
clf.score(X_test, y_test)
```

**TASK 3**

**RANDOM FOREST AND HYPERPARAMETER TUNING**

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
n_estimators=[int(x) for x in np.linspace(start=10, stop=80, num=80)]
max_features=['auto', 'sqrt']
max_depth=[2,4]
min_samples_split=[2,5]
min_samples_leaf=[1,2]
bootstrap=[True, False]
```

```python
param_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf,
               'bootstrap': bootstrap}
```

```python
rf_Model=RandomForestClassifier()
from sklearn.model_selection import GridSearchCV
rf_Grid = GridSearchCV(estimator = rf_Model, param_grid = param_grid, cv = 3, verbose=2, n_jobs = 4)
```

```python
rf_Grid.fit(X_train, y_train)
```

```python
rf_Grid.score(X_test, y_test)
```
